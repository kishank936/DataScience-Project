```python
#Read an Image

img = cv2.imread('dog.jfif')

newimg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
plt.show()
plt.imshow(newimg)
plt.show()
```

```python
import cv2

img = cv2.imread('dog.jfif')

cv2.imshow('Dog Image', img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#img change color

import cv2

img = cv2.imread('dog.jfif')
gray = cv2.imread('dog.jfif' , cv2.IMREAD_GRAYSCALE)

cv2.imshow('Dog Image', img)
cv2.imshow('Gray Dog Image',gray)

cv2.waitKey(0)
cv2.destroyAllWindows()

```
